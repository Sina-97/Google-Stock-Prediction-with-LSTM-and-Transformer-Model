# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.seasonal import STL
import scipy.stats
```

# Reading

```python
path = '../data/GOOGL_2006-01-01_to_2018-01-01.csv'
df = pd.read_csv(path)
df.head()
```

# Preprocessing

```python
df['Date'].iloc[0]
# Date is in string format
```

```python
# Converting Date column to datetime
df['Date'] = pd.to_datetime(df.Date, yearfirst=True)
df.head()
```

```python
# Checking if data is equally spaced
df['Date']
difference = df['Date'].iloc[1:].reset_index(drop=True) - df['Date'].iloc[:-1].reset_index(drop=True)

plt.subplots(figsize=(10,3))
plt.scatter(df['Date'].iloc[1:], pd.DataFrame(difference)['Date'].dt.days, s=2)
plt.xlabel("Sample's date")
plt.ylabel('Days from last sample')
plt.title("Checking if data is equally spaced")
plt.show()
```

```python
df.set_index('Date', inplace=True)
```

```python
# Setting the frequency to business days
df = df.asfreq('b')
```

```python
df['Name'].unique()
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
# Checking for null values
df.isna().sum()
```

```python
# Filling missing values
df = df.fillna(method='ffill') # Frontfill
```

```python
df.head()
```

# Visualization

```python
# Distributions for different variables
for column in df.columns[:-1]:
    aux_df = df[column]
    plt.plot(aux_df.index, aux_df, label=str(column))

plt.ylabel("Currency")
plt.xlabel("Date")
plt.title("Distribution for Open, High, Low and Close")
plt.legend()
plt.show()
```

```python
plt.subplots(figsize=(10, 3))
plt.plot(df.index, df['Volume'])
plt.ylabel("Volume")
plt.xlabel("Date")
plt.title("Volume Distribution across the years")
plt.show()
```

```python
plt.subplots(figsize=(10, 3))

filter_1 = df[df.index.year == 2010]
filter_2 = filter_1[filter_1.index.month == 6]

for column in df.columns[:-1]:
    aux_df = filter_2[column]
    plt.plot(aux_df.index, aux_df, 'o-', label=str(column))

plt.ylabel("Currency")
plt.xlabel("Date")
plt.title("Distribution for Open, High, Low and Close for May 2010")
plt.legend()
plt.show()
```

# Extracting Trends

```python
for column in df.columns:
    aux_series = df[column]

    stl = STL(aux_series, period= 365-(52+53)) # taking out the weekends of a year
    res = stl.fit()
    fig = res.plot()
    plt.show()
```

```python
for column in df.columns:
    aux_series = df[column]

    stl = STL(aux_series, period=365-(52+53))
    res = stl.fit()

    # Removing trends
    diff = pd.DataFrame(aux_series - res.trend, columns=['Diff'])
    # Aggregating by month and days
    diff['Month'] = diff.index.month
    diff['Day'] = diff.index.day
    
    fig, ax = plt.subplots(1, 2, figsize = (10, 6))

    sns.boxplot(x='Diff', y='Month', data=diff, ax = ax[1], orient='h')
    ax[1].set_title('Boxplots Grouped by Month')

    sns.boxplot(x='Diff', y='Day', data=diff, ax = ax[0], orient='h')
    ax[0].set_title('Boxplots Grouped by Day')

    plt.suptitle("Checking seasonality for:"+column)
    plt.show()
```

# Checking distributions

```python
for col in df.columns:
    scipy.stats.probplot(df[col], plot=plt)
    plt.title(col)
    plt.show()
```
